# 1.0 Data Preparation 

```python
import pandas as pd
```

```python
data = pd.read_csv('Hitters.csv')
```

pre-process the data

```python
# remove all rows have missing values on any variable
data = data.dropna()
data = data.iloc[:,1:]
```

```python
#just keep numetic data
data = data.drop(['League','Division','NewLeague'],axis = 1)
```

# 1.1 Lasso Regression

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = data['Salary']
x = data.drop(['Salary'], axis = 1)
```

```python
# use cross-validation to get the optimal penalty value
from sklearn.linear_model import LassoCV
alphas = 10**np.linspace(-10,10,100)
lassocv = LassoCV(alphas = alphas, normalize=True, cv = 10, max_iter=10000)
lassocv.fit(x, y)
best_alpha = lassocv.alpha_
print('the best penalty from cross-validation is：' + str(best_alpha))

#use the best alpha to build lasso regression
lasso = Lasso(alpha = best_alpha, normalize=True, max_iter=10000)
lasso.fit(x,y)
coef = lasso.coef_
#get the number of predictors which isn't zero
left_predictor = [p for p in coef if p != 0]
print('the number of remaing predictors are:',len(left_predictor))

```

```python
alphas = 10**np.linspace(-10,10,100)
lasso_cofficients = []
 
for alpha in alphas:
    lasso = Lasso(alpha = alpha, normalize=True, max_iter = 10000)
    lasso.fit(x, y)
    lasso_cofficients.append(lasso.coef_)
    
  

plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (20,10))
plt.style.use('ggplot')
plt.plot(alphas, lasso_cofficients)
plt.xscale('log')
plt.axis('tight')
plt.title(r'alpha coefficient trajectories')
plt.xlabel('Log Alpha')
plt.ylabel('Cofficients')
plt.legend(labels=x.columns)
plt.show()
```

according to the coefficient plot and coefficient value, we can know that the last 3 predictor s remain in the model are Hits, Walks and CRBI

# 1.2 Ridge Regression

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
```

```python
# use cross-validation to get the optimal penalty value
alphas = 10**np.linspace(-10,10,100)
ridgecv = RidgeCV(alphas = alphas, normalize=True, scoring='neg_mean_squared_error', cv = 10)
ridgecv.fit(x, y)
best_alpha = ridgecv.alpha_
print('the best penalty from cross-validation is：' + str(best_alpha))

#use the best alpha to build lasso regression
ridge = Ridge(alpha = best_alpha, normalize=True)
ridge.fit(x,y)
coef = ridge.coef_
#get the number of predictors which isn't zero
print('coefficients: ',coef)
left_predictor = [p for p in coef if p != 0]
print('the number of remaining predictors are:',len(left_predictor))
```

```python
alphas = 10**np.linspace(-10,10,100)
ridge_cofficients = []
 
for alpha in alphas:
    ridge = Ridge(alpha = alpha, normalize=True)
    ridge.fit(x, y)
    ridge_cofficients.append(ridge.coef_)
    
 
plt.figure(figsize = (20,10))
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('ggplot')
plt.plot(alphas, ridge_cofficients)
plt.xscale('log')
plt.axis('tight')
plt.title(r'alpha coefficient trajectories')
plt.xlabel('Log Alpha')
plt.ylabel('Cofficients')
plt.legend(labels=x.columns)
plt.show()
```

# 2

Bias is the difference between your prediction and the value you want to predict. 
Variance is the amount that the estimate of the target function will change if different training data was used. 

When your model is complex and have lots of parameters, you will capture the specifics of training dataset and have lower bias. But at this time, your model is highly variable, because any tiny change in training dataset would change the number and type of paramters.

When your model is simple and just have a few parameters, you won't capture the specifics of training dataset very well, so that you would have a high bias.But at this time, your model is not very variable, because the small change in training dataset won't change you model a lot.

Regularizaiton helps to reduce number of parameters in the model, so that helps to reduce variance at the cost of higher bias. The ridge model have higher variance but lower bias, because all the parameters are still in the model. The lasso model have lower variance but higher bias, becasue some parameters shrinked to zero, which made the model simpler.
